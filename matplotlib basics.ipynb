```python
import matplotlib.pyplot as plt
import numpy as np
```

### Line chart

plt.plot(x,y)# Plot y versus x as lines and/or markers

```python
x=np.linspace(0,10,20)
y=x*x
```

```python
plt.plot(x,y)
```

## Add lables and titles

```python

plt.plot(x,y)

## Adding labels
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

## Adding Title to the plot
plt.title('Y vs X')
```

### Formating color

```python
plt.plot(x,y)

## Adding labels
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

## Adding Title to the plot
plt.title('Y vs X')
## Formating a line  color
plt.plot(x,y,color='r')
```

```python
x=np.linspace(0,15,20)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

## Adding Title to the plot
plt.title('Y vs X')
## Formating a line  color
plt.plot(x,y,color='r')
```

### Formating a line style

```python
plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

## Adding Title to the plot
plt.title('Y vs X')

## Adding color and line style
plt.plot(x,y,color='g',linestyle='--')
```

## Formating line width and Markers

```python
plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

## Adding Title to the plot
plt.title('Y vs X')
## Adding color,linestyle and line width and markers
plt.plot(x,y,color='g',linestyle='--',linewidth=3,marker='o')
```

## Format Strings

```python
plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

## Adding Title to the plot
plt.title('Y vs X')
## Adding color,linestyle and marker
plt.plot(x,y,'g--o')
```

## Formating a Marker size

```python
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

## Adding Title to the plot
plt.title('Y vs X')

## Adding color,linestyle,marker and Markersize
plt.plot(x,y,'g--o',markersize=10)
```

## Formating and control transperancy of a chart

```python
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

## Adding Title to the plot
plt.title('Y vs X')

## Adding color,linestyle,marker,markersize and alpha
plt.plot(x,y,'g--o',markersize=8,alpha=0.7) # alpha lies in btw 0 to 1
```

## Adding legends

```python
# Prepare data
x=np.linspace(0,20,20)
y=x*x
z=x*x*x

# x vs y plot
plt.plot(x,y)

#x vs z plot
plt.plot(x,z)
```

```python
# Prepare data
x=np.linspace(0,20,20)
y=x*x
z=x*x*x

# x vs y plot
plt.plot(x,y,label='X-square')
#x vs z plot
plt.plot(x,z,label='X-Cube')
# Adding legend
plt.legend()
```

```python
# Prepare data
x=np.linspace(0,20,20)
y=x*x
z=x*x*x

# x vs y plot
plt.plot(x,y,label='X-square')
#x vs z plot
plt.plot(x,z,label='X-Cube')
# Adding legend,location of legend and fontsize
plt.legend(loc='best',fontsize='14')
plt.title('X')
```

Bar Plot

plt.bar(x,height,width=0.8,align='center')

```python
# Prepare data
x = ['python','R','C','C++','Java']
y = [10,8,4,7,9]
# Simple Bar Plot
plt.bar(x,height=y)
# Adding labels and title
plt.xlabel('Languages')
plt.ylabel('Rating')
plt.title('Rating for Languages')
```

```python
# Prepare data
x = ['python','R','C','C++','Java']
y = [10,8,4,7,9]
# Simple Bar Plot
plt.bar(x,height=y,width=0.2,alpha=0.7)
# Adding labels and title
plt.xlabel('Languages')
plt.ylabel('Rating')
plt.title('Rating for Languages')
```

```python
# Prepare data
x = ['c programing','java programing','python programing','c++ programming','R programming']
y=[10,5,15,4,3]
plt.barh(x,y)

## Adding title
plt.title('Horizontal Bar Chart')
```

```python
# Prepare data
x = ['c programing','java programing','python programing','c++ programming','R programming']
y=[10,5,15,4,3]
plt.barh(x,y,align='center',color='r',alpha=0.2)

## Adding title
plt.title('Horizontal Bar Chart')
```

## Histogram

plt.hist(x,bins=None)

Data Distribution of numerical variable

```python
# prepare data
x=[20,21,21,20,24,26,23,20,22,24,25,26,27,21,23,23]
plt.hist(x,bins=8) # bins =no.of cls intervals

# add lables and title
plt.xlabel('class interval')
plt.ylabel('frequency')
plt.title('histogram')
```

## PIE-CHART

plt.pie(x,explode=None,labels=) explode=slicing

```python
labels = ['python','R','C','C++','Java']
x = [10,8,4,7,9]
# simple pie chart
plt.pie(x,labels=labels)
```

```python
labels = ['python','R','C','C++','Java']
x = [10,8,4,7,9]
# simple pie chart
plt.pie(x,labels=labels,explode=[0.2,0.2,0.4,0.2,0.3])
```
