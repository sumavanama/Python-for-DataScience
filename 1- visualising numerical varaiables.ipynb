## visualising the distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Import House Sales Data
df=pd.read_excel('housesales.xlsx')
df
```

## Plotting Univaraiable Distribution

```python
sns.distplot(df['listprice'])
```

```python
sns.distplot(df['listprice'],hist=False)
```

```python
sns.distplot(df['listprice'],hist=False,rug=True)
```

```python
sns.distplot(df['listprice'],color='r')
```

```python
sns.distplot(df['listprice'],hist_kws={'color':'g'},kde_kws={'color':'r'})
```

```python
from scipy import stats
```

```python
sns.distplot(df['listprice'],fit=stats.norm)
```

```python
# create a distribution plot for square feet
sns.distplot(df['sqft'])
```

## Plotting Bi-variate

```python
## Relationship btw sqft and list price
```

```python
sns.jointplot(x='sqft',y='listprice',data=df)
```

```python
sns.jointplot(x='sqft',y='listprice',data=df,kind='hex')
```

```python
sns.jointplot(x='sqft',y='listprice',data=df,kind='kde')
```

```python
# create a jointtplot for days on market and list price
sns.jointplot(x='daysonmarket',y='listprice',data=df)
```

## Pair wise Relationship

```python
Numerical=df[['sqft','listprice','daysonmarket','beds','lotsize','hometype','parkingtype']]
Numerical.head(2)
```

```python
sns.pairplot(data=Numerical)
```

```python
subset=df[['sqft','listprice','daysonmarket','beds','lotsize','hometype','parkingtype']]
subset.head(2)
```

```python

sns.pairplot(data=subset,hue='hometype')
```

```python
# Select only sqft,daysonmarket, listprice and hometype
Numerical=df[['sqft','daysonmarket','listprice','hometype']]
sns.pairplot(data=Numerical,hue='hometype')
```

```python
Numerical=df[['sqft','daysonmarket','listprice','hometype']]
sns.pairplot(data=Numerical,hue='hometype',markers=['o','+','<'])
```

```python
sns.pairplot(data=df,vars=['sqft','lotsize','listprice'],hue='parkingtype')
```
