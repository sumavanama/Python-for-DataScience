## Categorical distribution

# Boxplot

**`sns.boxplot(x,y,hue,data)`**

```python
sns.boxplot(x='hometype',y='listprice',data=df)
```

```python
sns.boxplot(x='hometype',y='listprice',hue='parkingtype',data=df)
```

```python
sns.boxplot(x='hometype',y='sqft',hue='parkingtype',data=df)
```

```python
sns.boxplot(y='zip',x='listprice',orient='h',data=df)
```

## Violinplot

**`sns.violinplot(x,y,hue,data)`**

```python
sns.violinplot(x='hometype',y='listprice',data=df)
```

```python
sns.violinplot(x='hometype',y='listprice',hue='parkingtype',data=df)
```

```python
sns.violinplot(x='hometype',y='listprice',hue='parkingtype',data=df,split=True)
```

```python
sns.violinplot(x='zip',y='listprice',hue='hometype',data=df)
```

```python
sns.violinplot(x='zip',y='listprice',hue='hometype',data=df,scale='count',inner='quartile')
```

```python
sns.boxenplot(x='hometype',y='listprice',data=df) 
```

## Categorical Estimates

**`sns.barplot(x,y,hue,data,ci)`**

```python
sns.barplot(x='hometype',y='listprice',data=df)
```

```python
# calculate average listprice for single family,town,condo

df.groupby('hometype')['listprice'].mean()
```

```python
sns.barplot(x='zip',y='listprice',hue='hometype',data=df)
```

```python
# create a visualization to understand avg listprice fr each hometype and beds

sns.barplot(x='hometype',y='listprice',hue='beds',data=df)
```

```python
sns.barplot(x='hometype',
            y='listprice',
            hue='beds',
            data=df,
            estimator=np.median)
```

```python
sns.barplot(x='hometype',
            y='listprice',
            hue='beds',
            data=df,
            ci='sd')
```

```python
sns.barplot(x='hometype',y='listprice',ci=95,data=df)
```
